# CMS dimuon invariant mass exercise

This notebook make use of a subset of 2010 CMS collision data released a Open Data:

* McCauley, T. (2014). Dimuon event information derived from the Run2010B public Mu dataset. CERN Open Data Portal. DOI: [10.7483/OPENDATA.CMS.CB8H.MFFA](http://opendata.cern.ch/record/700).

The file we are using contains about 10000 $\mu\mu$ events recorded in 2010. Each line corresponds to an event, and present the kinematical information about the recorded muon pair:

`Run,Event,E1,px1 ,py1,pz1,pt1,eta1,phi1,Q1,E2,px2,py2,pz2,pt2,eta2,phi2,Q2,M`

Each muon candidate is characterised by:

- `E`: energy
- `px,py,pz`: momentum components
- `pt,eta,phi`: tranverse momentum, pseudorapidity, azimuthal angle
- `Q`: electrical charge

## 0. Read in the data into a ROOT `TTree`

This step might take a while: be patient!

```python
import ROOT

inputFileName = "MuRun2010B_Tutorial.csv"

dimuons = ROOT.TTree("MuonPairs","MuonPairs")
dimuons.ReadFile(inputFileName)
#dimuons.Print()
```

## 1. Plot the dimuon invariant mass spectrum using `TTree::Draw`

We begin by creating an histogram to hold the invariant mass values. 

In case you are not familiar with ROOT, the [manual](https://root.cern.ch/doc/master/index.html) is your friend. For instance, here's the entry for the [`TH1F`](https://root.cern.ch/doc/master/classTH1F.html) histogram object.

```python
m_mumu_1 = ROOT.TH1F("m_mumu_1","CMS Opendata: #mu#mu invariant mass;m_{#mu#mu} [GeV];Events/GeV",240, 2, 122)
```

In order to loop on the TTree rows, we can use the [TTree::Draw](https://root.cern.ch/doc/master/classTTree.html) method: this is the most straightforward way to loop over the content a ROOT ntuple, even if not the most powerful for complex selections or data treatment. 

Let's define the formula for the invariant mass of the muon pair:

$$ m_{\mu\mu} = \sqrt{ (E_1 + E_2)^2 - (\vec{p}_1 + \vec{p}_2)^2}$$

```python
invMassFormula = "sqrt((E1 + E2)^2 - ((px1 + px2)^2 + (py1 + py2)^2 + (pz1 + pz2)^2))"
```

Let's then define a **selection** for the events. In the simple example below, we require that the muons in the pair have opposite electricla charge (e.g. they are a muon and an anti-muon):

```python
selection = "Q1*Q2==-1"
```

Now let's loop on the ntuple with the `TTree::Draw()` method and store the result into the `m_mumu_1` histogram using the `>>` operator, then plot it.

### Questions: 

* **What can you observe in the resulting invariant mass spectrum?**
* **How many peaks can you see?** 
* **What do they correspond to?**

```python
dimuons.Draw(invMassFormula + " >> m_mumu_1","","hist")

c = ROOT.TCanvas()

m_mumu_1.Draw("E")
m_mumu_1.SetMarkerStyle(20)
m_mumu_1.SetLineColor(1)
m_mumu_1.SetStats(0)

c.SetLogx() # Log scale on X axis
c.SetLogy() # Log scale on Y axis

c.Draw()
```

## 2. Plot the dimuon invariant mass spectrum using and explicit loop

Below there is an alternative method to **explicitely** loop on our `TTree` without using the `TTree::Draw()` method. When using this direct approach, it is easier the include specific selections on the data properties, to compute complex quantities, and to fill multiple histograms. The code below reproduces exactly what we did before, now using this new approach:

```python
from math import sqrt

m_mumu_2 = ROOT.TH1F("m_mumu_2","CMS Opendata: #mu#mu invariant mass;m_{#mu#mu} [GeV];Events/GeV",240, 2, 122)

for e in dimuons: # an esplicit loop on the events. "e" is an event
    if e.Q1 * e.Q2 != -1: # this is the "opposite charge" selection we had before
        continue
    # compute invariant mass
    m = sqrt( (e.E1 + e.E2)**2 - ((e.px1 + e.px2)**2 + (e.py1 + e.py2)**2 + (e.pz1 + e.pz2)**2) )
    m_mumu_2.Fill(m)
```

```python
c = ROOT.TCanvas()

m_mumu_2.Draw("E")
m_mumu_2.SetMarkerStyle(20)
m_mumu_2.SetLineColor(1)
m_mumu_2.SetStats(0)

c.SetLogx() # Log scale on X axis
c.SetLogy() # Log scale on Y axis

c.Draw()
```
