```python
print("Creating a simple histogram...", end=" ")
h = ROOT.TH1F("h","",100,-5.,5.)
print("done.")

print("Fillin it with 10000 random values...", end=" ")
h.FillRandom("gaus",10000)
print("done.")

print("And plotting it...", end=" ")
c = ROOT.TCanvas("c","",800,600)
h.Draw()
c.Draw()
print("done.")
```
